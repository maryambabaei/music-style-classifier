```python

import pandas as pd
file_path1 = '/content/drive/MyDrive/DJ Project /CSV_files/Music Data Set 1fbeabcb4aad8014830af8919a1d5357_all.csv'
file_path2 = '/content/drive/MyDrive/DJ Project /CSV_files/music_features.csv'
file_path3 = '/content/drive/MyDrive/DJ Project /CSV_files/download_log.csv'

try:
    df1 = pd.read_csv(file_path1)
    print("Successfully read:", file_path1)
    print(df1.head()) # Print the first few rows to confirm

except FileNotFoundError:
    print(f"Error: File not found at {file_path1}")
except Exception as e:
    print(f"An error occurred while reading {file_path1}: {e}")

try:
    df2 = pd.read_csv(file_path2)
    print("\nSuccessfully read:", file_path2)
    print(df2.head()) # Print the first few rows to confirm
except FileNotFoundError:
    print(f"Error: File not found at {file_path2}")
except Exception as e:
    print(f"An error occurred while reading {file_path2}: {e}")

try:
    df3 = pd.read_csv(file_path3)
    print("\nSuccessfully read:", file_path3)
    print(df3.head()) # Print the first few rows to confirm
except FileNotFoundError:
    print(f"Error: File not found at {file_path3}")
except Exception as e:
    print(f"An error occurred while reading {file_path3}: {e}")
```

```python
# prompt: concat Artist and TItle in df1 columns and replace it with thees two columns

df1['Artist - Title'] = df1['Artist'] + ' - ' + df1['Title']
df1 = df1.drop(columns=['Artist', 'Title'])

print(df1.head())
```

```python
# prompt: i have three dataframes, df 1 conatains labels for 'Artist - Title' column df2 conatains features for the musics and df3 cantains relation between these two dataframe as 'Search_Title' is the 'Artist-Title' column in df1 and ' Downloaded_Filename'  is 'filename' in df2, i want to merge df1 and df2 based on similarty in df3

import pandas as pd
# Merge df3 with df1 on 'Search_Title' and 'Artist - Title'
merged_df = pd.merge(df3, df1, left_on='Search_Title', right_on='Artist - Title', how='inner')

# Merge the result with df2 on 'Downloaded_Filename' and 'filename'
final_df = pd.merge(merged_df, df2, left_on='Downloaded_Filename', right_on='filename', how='inner')

# Drop redundant columns if necessary (e.g., 'Search_Title', 'Artist - Title', 'filename' from merged_df)
final_df = final_df.drop(columns=['Artist - Title', 'filename'])

print("\nFinal Merged DataFrame:")
print(final_df.head())
print("\nShape of the final merged DataFrame:", final_df.shape)
```

```python
# prompt: print final_df column names

final_df.columns
```

```python
# prompt: remove , 'Downloaded_Filename', 'Status', 'Method_Used',
#        'Source_URL', 'Artist_Info', 'Download_Date', 'File_Size_MB',
#        'File_Path','Library Tag','duration' columns from df

cols_to_drop = ['Downloaded_Filename', 'Status', 'Method_Used',
       'Source_URL', 'Artist_Info', 'Download_Date', 'File_Size_MB',
       'File_Path','Library Tag','duration']

final_df = final_df.drop(columns=cols_to_drop)

print("\nDataFrame after dropping columns:")
print(final_df.head())
print("\nShape of the final merged DataFrame:", final_df.shape)
final_df.columns
```

```python
# prompt: save final_df as music_df in Dj project folder

# Define the file path for saving the dataframe
output_path = '/content/drive/MyDrive/DJ Project /CSV_files/music_df.csv'

# Save the final_df to a CSV file
final_df.to_csv(output_path, index=False)

print(f"\nFinal DataFrame saved to: {output_path}")
```
