```python
# prompt: drop rows that contain nan values

df_dropped = df.dropna()

print("\nOriginal DataFrame shape:", df.shape)
print("DataFrame shape after dropping rows with NaNs:", df_dropped.shape)

print("\nDataFrame head after dropping rows with NaNs:")
print(df_dropped.head())

df = df_dropped # Optionally replace the original DataFrame with the one without NaNs
```
